```python
import json 
def open_log_file(file):
    with open(file, 'r') as f:
        data = json.load(f)
    return data
```

```python
len(open_log_file(s0)) + len(open_log_file(s1)) + len(open_log_file(s2))

debate_answers = open_log_file(s0) + open_log_file(s1) + open_log_file(s2)
baseline_answers = open_log_file(b0) + open_log_file(b1) + open_log_file(b2)

assert len(debate_answers) == len(baseline_answers)
```

```python
import pandas as pd 
import os 
def compile_in_hallusion_format():
    hallusion_path = '/data4/saaket/hallusion_bench/'
    data_file = os.path.join(hallusion_path, 'HallusionBench.json')
    with open(data_file, 'r') as f:
        data = json.load(f)
    df = pd.DataFrame(data)

    # drop entries from 1055 to 1065 (both inclusive) 
    df = df.drop(df.index[1054:1065])
    print(len(df))
    # check if length of df is equal to length of debate_answers
    debate_answers_df = df.copy()
    baseline_answers_df = df.copy()

    debate_answers_df['model_prediction'] = debate_answers
    baseline_answers_df['model_prediction'] = baseline_answers

    debate_answers_df = debate_answers_df.drop(debate_answers_df.index[1053:1054])
    baseline_answers_df = baseline_answers_df.drop(baseline_answers_df.index[1053:1054])

    # make dirs
    os.makedirs('output/gpt4v/debate_baseline', exist_ok=True)
    os.makedirs('output/gpt4v/debate', exist_ok=True)

    debate_answers_df.to_json('output/gpt4v/debate/HallusionBench_result.json', orient='records')
    baseline_answers_df.to_json('output/gpt4v/debate_baseline/HallusionBench_result.json', orient='records')
```

```python
compile_in_hallusion_format()
```
