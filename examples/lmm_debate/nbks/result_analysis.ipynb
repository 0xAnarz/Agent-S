```python
def compute_metrics(ground_truth, responses):
    total = len(ground_truth)
    correct = sum(1 for gt, resp in zip(ground_truth, responses) if gt == resp)
    incorrect = sum(1 for gt, resp in zip(ground_truth, responses) if gt != resp and resp != "NA")
    unknown = sum(1 for resp in responses if resp == "NA")
    
    success_rate = correct / total
    error_rate = incorrect / total
    unknown_rate = unknown / total
    
    return {
        "Success Rate": success_rate,
        "Error Rate": error_rate,
        "Unknown Rate": unknown_rate
    }
```

```python
def compute_metrics(ground_truth, responses):
    total = len(ground_truth)
    correct = sum(1 for gt, resp in zip(ground_truth, responses) if gt == resp)
    incorrect = sum(1 for gt, resp in zip(ground_truth, responses) if gt != resp and resp != "NA")
    unknown = sum(1 for resp in responses if resp == "NA")
    
    success_rate = correct / total
    error_rate = incorrect / total
    unknown_rate = unknown / total
    
    return {
        "Success Rate": success_rate,
        "Error Rate": error_rate,
        "Unknown Rate": unknown_rate
    }
```
